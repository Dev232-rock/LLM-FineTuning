````python
import google.generativeai as genai
import json
import os

genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

model = genai.GenerativeModel("gemini-flash-latest")

FIELDS_SCHEMA = {
    "policy_number": "",
    "policyholder_name": "",
    "date": "",
    "time": "",
    "location": "",
    "description": "",
    "estimated_damage": "",
    "claim_type": ""
}

def gemini_extract_fields(text):
    prompt = f"""

{json.dumps(FIELDS_SCHEMA, indent=2)}

FNOL Text:
{text[:6000]}
"""
    response = model.generate_content(prompt)
    cleaned_response_text = response.text.strip()
    if cleaned_response_text.startswith('```json') and cleaned_response_text.endswith('```'):
        cleaned_response_text = cleaned_response_text[7:-3].strip()

    try:
        return json.loads(cleaned_response_text)
    except json.JSONDecodeError as e:
        print(f"JSONDecodeError: {e}")
        print(f"Model response.text: '{response.text}'")
        return dict(FIELDS_SCHEMA)

llm_fields = gemini_extract_fields(fnol_text)
llm_fields
````

````python
import re

def regex_extract(text):
    patterns = {
        "policy_number": r"POLICY NUMBER[:\s]+([A-Z0-9-]+)",
        "policyholder_name": r"NAME OF INSURED.*?:\s*(.+)",
        "date": r"DATE OF LOSS.*?(\d{2}/\d{2}/\d{4})",
        "location": r"LOCATION OF LOSS.*?:\s*(.+)",
        "estimated_damage": r"ESTIMATE AMOUNT.*?:\s*\$?([\d,]+)"
    }

    extracted = {}
    for field, pattern in patterns.items():
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            extracted[field] = match.group(1).strip()
    return extracted

regex_fields = regex_extract(fnol_text)
regex_fields
````

````python
MANDATORY_FIELDS = [
    "policy_number",
    "policyholder_name",
    "date",
    "location",
    "description",
    "claim_type",
    "estimated_damage"
]

def confidence_score(field, value, source):
    if not value:
        return 0.0

    base = 0.9 if source == "gemini" else 0.75
    if field in ["policy_number", "date", "estimated_damage"]:
        base += 0.05

    return min(base, 0.99)

def merge_fields(llm, regex):
    final = {}

    for field in MANDATORY_FIELDS:
        if llm.get(field):
            final[field] = {
                "value": llm[field],
                "confidence": confidence_score(field, llm[field], "gemini")
            }
        elif regex.get(field):
            final[field] = {
                "value": regex[field],
                "confidence": confidence_score(field, regex[field], "regex")
            }

    return final

extracted_fields = merge_fields(llm_fields, regex_fields)
extracted_fields
````

````python
def find_missing_fields(fields):
    return [f for f in MANDATORY_FIELDS if f not in fields]

missing_fields = find_missing_fields(extracted_fields)
missing_fields
````

````python
def route_claim(fields, missing_fields):
    description = fields.get("description", {}).get("value", "").lower()
    damage = int(fields.get("estimated_damage", {}).get("value", "0").replace(",", ""))

    if missing_fields:
        return "Manual Review", "Mandatory fields are missing"

    if any(word in description for word in ["fraud", "staged", "inconsistent"]):
        return "Investigation Flag", "Suspicious keywords detected"

    if fields.get("claim_type", {}).get("value") == "injury":
        return "Specialist Queue", "Injury-related claim"

    if damage < 25000:
        return "Fast-track", "Low estimated damage"

    return "Standard Processing", "Default routing"

route, reason = route_claim(extracted_fields, missing_fields)
route, reason
````

````python
final_output = {
    "extractedFields": extracted_fields,
    "missingFields": missing_fields,
    "recommendedRoute": route,
    "reasoning": reason
}

print(json.dumps(final_output, indent=2))
````

````python
LABEL_WORDS = [
    "policy", "insured", "contact", "description",
    "location", "date", "time", "number"
]

def is_label(value):
    v = value.lower()
    return any(word in v for word in LABEL_WORDS) and len(v.split()) < 6
````

````python
def merge_fields(llm, regex):
    final = {}

    for field in MANDATORY_FIELDS:
        llm_val = llm.get(field, "").strip()
        regex_val = regex.get(field, "").strip()

        if llm_val and not is_label(llm_val):
            final[field] = {
                "value": llm_val,
                "confidence": confidence_score(field, llm_val, "gemini")
            }
        elif regex_val:
            final[field] = {
                "value": regex_val,
                "confidence": confidence_score(field, regex_val, "regex")
            }

    return final
````

````python
def confidence_score(field, value, source):
    if not value or is_label(value):
        return 0.3

    base = 0.9 if source == "gemini" else 0.75
    if field in ["policy_number", "date", "estimated_damage"]:
        base += 0.05

    return min(base, 0.99)
````
